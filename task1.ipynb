 # Task 1: Prediction using Supervised ML

 In this regression task we will predict the percentage of marks that a student is expected to score based upon
the number of hours they studied. This is a simple linear regression task as it involves just two variables.





# Steps to follow for prediction using Supervised ML are:
 
 Step 1: Importing all the libraries.

 Step 2: Importing all the dataset.

 Step 3: Visualising the dataset.

 Step 4: Data preparation.

 Step 5: Training the algorithm.

 Step 6: Visualising the model.

 Step 7: Making predictions
 
 Step 8: Evaluating the model.

# Step 1: Importing all the header files

```python

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Step 2: Importing all the dataset

```python
url = "http://bit.ly/w-data"
sample_data = pd.read_csv(url)
print("Data imported successfully")
sample_data.head()
```

# Step 3: Visualising the dataset

Let's plot our data points on 2-D graph to eyeball our dataset and see if we can manually find any relationship
between the data. We can create the plot with the following script:

```python
sample_data.plot(x='Hours', y='Scores',style='o', markersize=7, marker=r'$\clubsuit$')
plt.title('Hours vs percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.tight_layout()
plt.show()
```

From the graph above, we can clearly see that there is a positive linear relation between the number of hours
studied and percentage of score.


# Step 4: Data preparation

The next step is to divide the data into "attributes" (inputs) and "labels" (outputs).


```python
X = sample_data.iloc[:, :-1].values
y = sample_data.iloc[:, 1].values
```

Now that we have our attributes and labels, the next step is to split this data into training and test sets. We'll do
this by using Scikit-Learn's built-in train_test_split() method:

# Step 5: Training the algorithm

We have to split our data into training and testing sets, and now is finally the time to train our algorithm.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print("Training complete.")
```

# Step 6: Visualising the model

```python
# Plotting the regression line
line = model.coef_*X + model.intercept_
# Plotting for the test data
plt.scatter(X_train, y_train, color='red')
plt.plot(X, line, color='blue');
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

# Step 7: Making prediction

Now that we have trained our algorithm, it's time to make some predictions.

```python

y_pred = model.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
# You can also test with your own data
hours = 9.25

own_pred = model.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

# Step 8: Evaluating the model

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how
well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square
error. There are many such metrics.


```python
# Evaluating the model
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```
